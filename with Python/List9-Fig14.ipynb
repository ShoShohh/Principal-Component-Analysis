```python
from sklearn.datasets import fetch_openml #Mnistデータのダウンロードに必要
import matplotlib.pyplot as plt

#Rではspamデータを使いましたが，Pythonでは有名なMnistデータ（文字画像）を使ってみます．
mnist = fetch_openml('mnist_784')
```

```python
#Mnistデータは784次元（28*28の画像）のデータが70000個ある．
mnist.data 
```

```python
#Mnistデータの各ラベル．各データが数字の何を表すのかのラベル．
mnist.target
```

```python
#number+1番目のデータが数字の何を表し，文字画像としてどんなデータなのか白黒で表示する．
number = 10
img = mnist.data.loc[number].values.reshape(28, 28)
plt.imshow(img, 'gray')
print(mnist.target.loc[number])
```
