```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(data = iris.data[50:100, :], columns = iris.feature_names)

scaler = StandardScaler()
scaler.fit(df) 
iris_scaled = scaler.transform(df)
```

```python
pca1 = PCA(n_components =1)
PCscore1 = pca1.fit_transform(iris_scaled)
PCInverse1 = pca1.inverse_transform(PCscore1)
print(PCInverse1)
```

```python
pca2 = PCA(n_components =2)
PCscore2 = pca2.fit_transform(iris_scaled)
PCInverse2 = pca2.inverse_transform(PCscore2)
print(PCInverse2)
```
